```python
# Array properties

arr = np.array([[1,2,3,4,],[0,3,4,5,]])
print(arr)

# shape of array
print(f' the shape of the array is {arr.shape}')

# size of array
print(f' the size of the array is {arr.size}')

#Datatype of array
print(f' the Data type of the array is {arr.dtype}')

# Dimensions of array
print(f' the dimension of the array is {arr.ndim}')
```

```python
# Indexing and sliding
# Access the elements in 1D array

arr = np.array([1,2,3,4])
print(arr)
print(arr[3])


# sliding the elements
print(arr[0:3])

#Access the elements in 2D array
arr2 = np.array([[1,2,3],[2,3,4]])
print(arr2)
print(arr2[1,1])
print(arr2[0,:])#frist row with all columns
print(arr2[:,0])#frist columns with all rows
```

```python
#mathermatical operations

a = np.array([1,2,3])
b = np.array([3,4,5,])

print(f' the sum of a and b is {a + b}')
print(f' the difference of a and b is {a - b}')
print(f' the product of a and b is {a * b}')
print(f' the division of a and b is {a / b}')
```

```python
b1 = np.array([[1,2,3],[2,3,4,]])
b2 = np.array([[5,5],[6,6],[4,4]])

#dot product _____ product = a
c = np.dot(b1,b2)
print(c)
```
